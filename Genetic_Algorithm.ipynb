```python
bridge_length = 100
min_range = 3 # x 값의 범위 최소
max_range = 7 # x 값의 범위 최대

min_xnum = 1 # 한 범위 내 x값 개수 최소
max_xnum = 3 # 한 범위 내 x값 개수 최대

max_y = 20
```

```python
# 무작위 다리 생성
def generate_bridge(range_num):
  range_length = bridge_length // range_num
  x_list = []
  y_list = []
  for x_range in range(range_num):
    x_num = random.randint(min_xnum, max_xnum)
    for x in range(x_num):
      x_val = random.randint(range_length * x_range, range_length * (x_range + 1))
      y_val = random.randint(0, max_y)
      x_list.append(x_val)
      y_list.append(y_val)

  return [x_list, y_list]
```

```python
# 첫번째 세대 생성
def generate_population(size, min_range, max_range):
  population = []
  for i in range(size):
    range_num = i % (max_range - min_range + 1) +  min_range
    population.append(generate_bridge(range_num))

  return population
```

```python
def compute_performance(population):
  performance_list = []
  for individual in population:
    score = fitness(individual)
    performance_list.append([individual, score])

  population_sorted = sorted(performance_list, key=lambda x: x[1], reverse=True)

  return population_sorted
```

```python
def select_survivors(population_sorted, best_sample, lucky_few):
  next_generation = []
  for i in range(best_sample):
    next_generation.append(population_sorted[i][0])

  lucky_survivors = random.sample(population_sorted, k=lucky_few)
  for l in lucky_survivors:
    next_generation.append(l[0])

  if len(next_generation) < best_sample + lucky_few:
    next_generation.append(generate_bridge(length=random.randint(min_range, max_range)))

  random.shuffle(next_generation)

  return next_generation
```

```python
pop_sorted = compute_performance(test_generation)
survivors = select_survivors(pop_sorted, best_sample=20, lucky_few=20)
```
